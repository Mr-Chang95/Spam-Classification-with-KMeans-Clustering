```python
def return_txt(msg):
    if msg.get_content_type()=='text/plain':
        return msg.get_payload()
    if msg.get_content_type()=='text/html':
        return "I BETTER ADD SOME CODE HERE"
    else:
        for item in msg.get_payload():
            return return_txt(item)
            break

```
